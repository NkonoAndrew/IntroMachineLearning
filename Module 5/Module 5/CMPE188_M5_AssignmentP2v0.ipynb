## Initialization

```python
# importing required libraries
import keras
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

## Creating a training set

```python
x_train = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
y_train = np.array([[0],[1],[1],[0]], "float32")
```

```python
x_train
```

```python
y_train
```

## Creating a simple neural network (same as our prediction example that we built from scratch -- here we use Python functions)

```python
model = Sequential()
model.add(Dense(3, input_dim = 2, activation = 'tanh'))
model.add(Dense(1, activation = 'tanh'))
```

```python
model.compile(loss = 'mean_squared_error') 
```

## Training the neural network

```python
history = model.fit(x_train, y_train, epochs = 550)
```

Using the neural network to predict the training data (we are training to measure just training accuracy at this point)

```python
y_pred = model.predict(x_train)
y_pred_binary = (y_pred > 0.5)*1
print('Predictions:\n', y_pred_binary)
print('Ground Truth\n', y_train)
print('Training Accuracy')
```
